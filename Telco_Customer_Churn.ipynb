```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the dataset

df = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
df.head()
```

```python
df.tail()
```

```python
# Check for missing values in the dataset
df.isnull()
```

```python
# Check for missing value count in the dataset
df.isnull().sum()
```

```python
# Drop unnecessary columns from the dataset
df = df.drop(['customerID','gender'], axis=1)
```

```python
df.head()
```

```python
# Check the data types of each column in the dataset
df.dtypes
```

```python
# Convert categorical columns to binary (0 and 1) for modeling
columns_to_binary = ['Partner', 'Dependents', 'PhoneService', 'PaperlessBilling', 'Churn']

for col in columns_to_binary:
    df[col] = df[col].map({'Yes': 1, 'No': 0})
```

```python
df.head()
```

```python
# Convert 'TotalCharges' column to numeric, coercing errors to NaN
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')

df['TotalCharges'] = df['TotalCharges'].fillna(0)
```

```python
# Convert categorical columns to dummy variables
df = pd.get_dummies(df, drop_first=True)
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.head()
```

```python
class_counts = df['Churn'].value_counts()
print(class_counts)
```

```python
# Plot the distribution of the target variable 'Churn'
import matplotlib.pyplot as plt
plt.hist(df['Churn'], bins=[-0.5,0.5,1.5], edgecolor='black')
plt.title('Churn Distribution')
plt.xlabel('Churn')
plt.ylabel('Frequency')
plt.xticks([0,1])
plt.show()
```

```python
# Split the dataset into features (X) and target variable (y)
X = df.drop(['Churn'],axis=1)
y = df['Churn']
```

```python
# Split the dataset into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2,stratify=y, random_state=42)
```

```python
# Scale the features using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
print("X_train_scaled shape:", X_train_scaled.shape) 
print("y_train shape:", y_train.shape)
```

```python
# Train a Random Forest Classifier with class weight balancing
from sklearn.ensemble import RandomForestClassifier

rf_model = RandomForestClassifier(class_weight='balanced_subsample',random_state=42)
rf_model.fit(X_train_scaled, y_train)
```

```python
rf_pred = rf_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy:", accuracy_score(y_test, rf_pred))
print("Precision:", precision_score(y_test, rf_pred))
print("Recall:", recall_score(y_test, rf_pred))
print("F1-Score:", f1_score(y_test, rf_pred))
print("ROC AUC Score:", roc_auc_score(y_test, rf_pred))
```

```python
# Train a Logistic Regression model with class weight balancing
from sklearn.linear_model import LogisticRegression

log_model = LogisticRegression(class_weight="balanced", solver="saga", max_iter=2000, random_state=42)
log_model.fit(X_train_scaled, y_train)
```

```python
log_pred = log_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy:", accuracy_score(y_test, log_pred))
print("Precision:", precision_score(y_test, log_pred))
print("Recall:", recall_score(y_test, log_pred))
print("F1-Score:", f1_score(y_test, log_pred))
print("ROC AUC Score:", roc_auc_score(y_test, log_pred))
```

```python
# Calculate the class weights for the target variable 'Churn'
churn_neg = class_counts[0]
churn_pos = class_counts[1]

print('Churn Positive:',churn_neg)
print('Churn Negative:',churn_pos)

scaled_pos_weight = churn_neg/churn_pos
print('Scaled Pos Weight:',scaled_pos_weight)
```

```python
# Train an XGBoost Classifier with class weight balancing
from xgboost import XGBClassifier

xgb_model = XGBClassifier(scale_pos_weight=2.77, max_depth = 5,eval_metric= 'aucpr', random_state=42)
xgb_model.fit(X_train_scaled, y_train)
```

```python
xgb_model_pred = xgb_model.predict(X_test_scaled)

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

print("Accuracy:", accuracy_score(y_test, xgb_model_pred))
print("Precision:", precision_score(y_test, xgb_model_pred))
print("Recall:", recall_score(y_test, xgb_model_pred))
print("F1-Score:", f1_score(y_test, xgb_model_pred))
print("ROC AUC Score:", roc_auc_score(y_test, xgb_model_pred))
```

```python
# Train a decision tree classifier
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(class_weight='balanced', random_state=42)
dt_model.fit(X_train_scaled, y_train)
```

```python
dt_pred = dt_model.predict(X_test_scaled)
print("Accuracy:", accuracy_score(y_test, dt_pred))
print("Precision:", precision_score(y_test, dt_pred))   
print("Recall:", recall_score(y_test, dt_pred))
print("F1-Score:", f1_score(y_test, dt_pred))
print("ROC AUC Score:", roc_auc_score(y_test, dt_pred))
```

```python
# Hyper parameter tuning in Random Forest
from sklearn.model_selection import GridSearchCV

rf_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [5, 10, None],
    'min_samples_split': [2, 5, 10]
}

rf_grid = GridSearchCV(estimator=rf_model, param_grid=rf_param_grid, cv=5, scoring='f1', n_jobs=-1)

print("Tuning Random Forest...")
rf_grid.fit(X_train_scaled, y_train)

print("Best Parameters:", rf_grid.best_params_)
print("Best Cross-Validation F1-Score:", rf_grid.best_score_)

best_rf = rf_grid.best_estimator_
```

```python
# Test predictions
rf_pred = best_rf.predict(X_test_scaled)
rf_prob = best_rf.predict_proba(X_test_scaled)[:, 1]

print("\n--- Tuned Random Forest Test Performance ---")
print("Accuracy:", accuracy_score(y_test, rf_pred))
print("Precision:", precision_score(y_test, rf_pred))
print("Recall:", recall_score(y_test, rf_pred))
print("F1-Score:", f1_score(y_test, rf_pred))
print("ROC AUC Score:", roc_auc_score(y_test, rf_prob))
```

```python
# Hyper parameter tuning in XGBoost
xgb_param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [3, 5, 7],
    'learning_rate': [0.01, 0.1, 0.2]
}

xgb_grid = GridSearchCV(estimator=xgb_model, param_grid=xgb_param_grid, cv=5, scoring='f1', n_jobs=-1)

print("Tuning XGBoost...")
xgb_grid.fit(X_train_scaled, y_train)

print("Best Parameters:", xgb_grid.best_params_)
print("Best Cross-Validation F1-Score:", xgb_grid.best_score_)

best_xgb = xgb_grid.best_estimator_
```

```python
# Test predictions
best_xgb = xgb_grid.best_estimator_
xgb_model_pred = best_xgb.predict(X_test_scaled)
xgb_model_prob = best_xgb.predict_proba(X_test_scaled)[:, 1]

print("\n--- Tuned XGBoost Test Performance ---")
print("Accuracy:", accuracy_score(y_test, xgb_model_pred))
print("Precision:", precision_score(y_test, xgb_model_pred))
print("Recall:", recall_score(y_test, xgb_model_pred))
print("F1-Score:", f1_score(y_test, xgb_model_pred))
print("ROC AUC Score:", roc_auc_score(y_test, xgb_model_prob))
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Calculate probabilities for the positive class (Churn = 1) for each model
rf_probs = rf_grid.best_estimator_.predict_proba(X_test_scaled)[:, 1]
lr_probs = lr_grid.best_estimator_.predict_proba(X_test_scaled)[:, 1]
xgb_probs = xgb_grid.best_estimator_.predict_proba(X_test_scaled)[:, 1]
dt_probs = dt_grid.best_estimator_.predict_proba(X_test_scaled)[:, 1]

# Compute ROC curve data (FPR, TPR) and AUC for each model
fpr_rf, tpr_rf, _ = roc_curve(y_test, rf_probs)
auc_rf = auc(fpr_rf, tpr_rf)

fpr_lr, tpr_lr, _ = roc_curve(y_test, lr_probs)
auc_lr = auc(fpr_lr, tpr_lr)

fpr_xgb, tpr_xgb, _ = roc_curve(y_test, xgb_probs)
auc_xgb = auc(fpr_xgb, tpr_xgb)

fpr_dt, tpr_dt, _ = roc_curve(y_test, dt_probs)
auc_dt = auc(fpr_dt, tpr_dt)

# Plot all ROC curves
plt.figure(figsize=(10, 7))

# Plot each model's curve with its respective AUC score in the label
plt.plot(fpr_xgb, tpr_xgb, color='darkorange', lw=2, label=f'XGBoost (AUC = {auc_xgb:.4f})')
plt.plot(fpr_rf, tpr_rf, color='blue', lw=2, label=f'Random Forest (AUC = {auc_rf:.4f})')
plt.plot(fpr_lr, tpr_lr, color='green', lw=2, label=f'Logistic Regression (AUC = {auc_lr:.4f})')
plt.plot(fpr_dt, tpr_dt, color='red', lw=2, label=f'Decision Tree (AUC = {auc_dt:.4f})')

# Plot the diagonal baseline representing a completely random model (no predictive power)
plt.plot([0, 1], [0, 1], color='gray', lw=1, linestyle='--', label='Random Guess (AUC = 0.5000)')

# Graph adjustments and styling
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate (FPR)', fontsize=12)
plt.ylabel('True Positive Rate (TPR)', fontsize=12)
plt.title('ROC Curves Comparison (Tuned Models)', fontsize=14, fontweight='bold')
plt.legend(loc="lower right", fontsize=11)
plt.grid(True, linestyle=':', alpha=0.6)

# Display the plot inside your notebook
plt.show()
```

```python
# Select the best model using auc score
models = ['Random Forest', 'Logistic Regression', 'XGBoost']
best_model = max(zip(models, [auc_rf, auc_lr, auc_xgb]), key=lambda x: x[1])
print(f"Best Model: {best_model[0]} with AUC = {best_model[1]:.4f}")
```
